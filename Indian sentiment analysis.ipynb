```python
import os
from nltk import tokenize
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import nltk

# Define the directory where speeches are stored
speeches_directory = r'D:\python_project\python_project\Indian_data'

# List of speeches with full paths
speeches = [
    'jan22.txt', 'feb22.txt', 'march22.txt', 'april22.txt', 'may22.txt', 'jun22.txt', 'july22.txt', 'aug22.txt', 'sept22.txt', 'oct22.txt', 'nov22.txt', 'dec22.txt',
    'jan2023.txt', 'feb2023.txt', 'march2023.txt', 'april2023.txt', 'may2023.txt', 'jun2023.txt', 'july2023.txt', 'aug2023.txt', 'sept2023.txt', 'oct2023.txt', 'nov2023.txt', 'dec2023.txt',
    'jan20.txt', 'feb20.txt', 'march20.txt', 'apr20.txt', 'may20.txt', 'jun20.txt', 'july20.txt', 'aug20.txt', 'sept20.txt', 'oct20.txt', 'nov20.txt', 'dec20.txt',
    'jan2021.txt', 'feb2021.txt', 'march2021.txt', 'april2021.txt', 'may2021.txt', 'jun2021.txt', 'jul2021.txt', 'aug2021.txt', 'sep2021.txt', 'oct2021.txt', 'nov2021.txt', 'dec2021.txt'
]
# Ensure NLTK resources are downloaded
nltk.download("punkt")
nltk.download("vader_lexicon")

score_dict = {}

# Initialize SentimentIntensityAnalyzer
sid_obj = SentimentIntensityAnalyzer()

# Iterate through each speech file
for speech in speeches:
    # Construct full path to the speech file
    speech_path = os.path.join(speeches_directory, speech)
    
    # Read the speech file with explicit encoding specification
    with open(speech_path, 'r', encoding='utf-8') as f:
        paragraph = f.read()

    # Tokenize the paragraph into sentences
    sentence_list = tokenize.sent_tokenize(paragraph)

    # Calculate average sentiment score for the speech
    sentenceSentimentSum = 0.0
    for sentence in sentence_list:
        sentiment_dict = sid_obj.polarity_scores(sentence)
        sentiment = sentiment_dict['compound']
        sentenceSentimentSum += sentiment

    # Calculate average sentiment score
    average_sentiment = sentenceSentimentSum / len(sentence_list)
    score_dict[speech] = round(average_sentiment, 4)

    # Print average sentiment score for the speech
    print(f"Average sentiment for {speech}: {round(average_sentiment, 4)}")

# Print the final score dictionary
print("Final scores:")
print(score_dict)
```

```python
df1 = pd.DataFrame(score_dict, index = [1]).T
df1.to_csv('indian_sentiment_scores.csv')
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd
import nltk
import os

# Function to perform sentiment analysis on a given text
def perform_sentiment_analysis(text):
    sid_obj = SentimentIntensityAnalyzer()
    sentiment_dict = sid_obj.polarity_scores(text)
    return sentiment_dict['compound']

# Function to analyze the sentiment impact on investor behavior, market mood, and public opinion
def analyze_sentiment_impact(speech_text):
    sentences = nltk.sent_tokenize(speech_text)
    investor_behavior = {"Positive": 0, "Negative": 0, "Neutral": 0}
    market_mood = 0
    public_opinion = 0

    for sentence in sentences:
        sentiment_score = perform_sentiment_analysis(sentence)

        # Analyze sentiment impact on investor behavior
        if sentiment_score > 0:
            investor_behavior["Positive"] += 1
        elif sentiment_score < 0:
            investor_behavior["Negative"] += 1
        else:
            investor_behavior["Neutral"] += 1

        # Aggregate sentiment scores for market mood and public opinion
        market_mood += sentiment_score
        public_opinion += sentiment_score

    # Calculate average sentiment for market mood and public opinion
    num_sentences = len(sentences)
    market_mood /= num_sentences
    public_opinion /= num_sentences

    return investor_behavior, market_mood, public_opinion

# Main function
def main():

    investor_behavior_list = []
    market_mood_list = []
    textfile_list = []

    for speech in speeches:
        speech_path = os.path.join(speeches_directory, speech)
        # Read the dataset
        with open(speech_path, 'r', encoding='utf-8') as f:
            speech_text = f.read()

        # Analyze sentiment impact
        investor_behavior, market_mood, public_opinion = analyze_sentiment_impact(speech_text)
        investor_behavior_list.append(investor_behavior)
        market_mood_list.append(market_mood)
        textfile_list.append(speech)
        print("added",speech)
        # Display results
        # print("Investor Behavior:")
        # print(investor_behavior)

        # print("\nMarket Mood:", market_mood)

        # print("\nPublic Opinion of Financial Institutions:", public_opinion)
    return textfile_list, investor_behavior_list, market_mood_list

if __name__ == "__main__":
    textfile_list, investor_behavior_list, market_mood_list = main()
```

```python

df_to_be_converted = pd.DataFrame({
    'textfile': textfile_list,
    'investor behavior': investor_behavior_list,
    'market mood': market_mood_list
})
df_to_be_converted.head()
```

```python
import pandas as pd

# Load the sentiment scores dataset
sentiment_scores_df = pd.read_csv('ind_eursenti.csv')

# Calculate the mean sentiment score
indian_mean_score = sentiment_scores_df['Indiansenti'].mean()

# Define thresholds for positive, negative, or neutral sentiment
positive_threshold = 0.1
negative_threshold = -0.1

# Assess the impact based on mean sentiment score
impact_on_investor_confidence = None
impact_on_market_stability = None
impact_on_economic_trends = None

# Example interpretation
if indian_mean_score > positive_threshold:
    impact_on_investor_confidence = "Positive"
    impact_on_market_stability = "Stable"
    impact_on_economic_trends = "Growth"
elif indian_mean_score < negative_threshold:
    impact_on_investor_confidence = "Negative"
    impact_on_market_stability = "Unstable"
    impact_on_economic_trends = "Downturn"
else:
    impact_on_investor_confidence = "Neutral"
    impact_on_market_stability = "Uncertain"
    impact_on_economic_trends = "Neutral or Uncertain"

# Print the assessment
print("Impact on Investor Confidence:", impact_on_investor_confidence)
print("Impact on Market Stability:", impact_on_market_stability)
print("Impact on Economic Trends:", impact_on_economic_trends)
```

```python
df_params = pd.DataFrame({
    "Impact on Investor Confidence":[impact_on_investor_confidence],
    "Impact on Market Stability": [impact_on_market_stability],
    "Impact on Economic Trends": [impact_on_economic_trends]
})

df_params.head()
```
